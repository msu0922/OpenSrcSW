```python
import csv

file = open('seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(file, delimiter = ',')

header = next(data) # header: ['날짜', '지점', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)']

tempDiff = 0
lowest = 999
highest = 0
date = [''] * 3

for row in data:
    if ((float(row[-1]) - float(row[-2])) >= tempDiff): # 일교차
        tempDiff = float(row[-1]) - float(row[-2])
        date[0] = row[0]
    if (float(row[-2]) <= lowest): # 최저기온
        lowest = float(row[-2])
        date[1] = row[0]
    if (float(row[-1]) >= highest): # 최고기온
        highest = float(row[-1])
        date[2] = row[0]

print('일교차가 가장 큰 날\n{0:s}\n최저 기온이 가장 낮은 날\n{1:s}\n최고 기온이 가장 높은 날\n{2:s}'.format(date[0], date[1], date[2]))

file.close()
```

```python
import csv

file = open('subway.csv', 'r', encoding = 'utf-8-sig')
temp = csv.reader(file, delimiter=',')
header = next(temp) # ['사용월', '호선명', '역ID', '지하철역', '승차승객수', '하차승객수']
line = {} # key = 호선명, value = 총 몇 개의 역이 있는지
for row in temp:
    if row[1] in line:
        line[row[1]] += 1
    else:
        line[row[1]] = 1

keys = list(line.keys())
values = list(line.values())
file.close()

file = open('subway.csv', 'r', encoding = 'utf-8-sig')
data = csv.reader(file, delimiter=',')
header = next(data) # ['사용월', '호선명', '역ID', '지하철역', '승차승객수', '하차승객수']

topStation = [''] * 3
top = [0] * 3
bottomStation = [''] * 3
bottom = [999999] * 3
for i in range(len(keys)): # 각 호선 별
    for j in range(values[i]): # 각 호선 별 한 줄씩
        rowList = next(data)
        total = int(rowList[-1]) + int(rowList[-2])
        # 상위 3개 찾기
        if (total >= top[0]):
            top[0] = total
            topStation[0] = rowList[-3]
        elif (total >= top[1]):
            top[1] = total
            topStation[1] = rowList[-3]
        elif (total >= top[2]):
            top[2] = total
            topStation[2] = rowList[-3]
        # 하위 3개 찾기
        if (total <= bottom[0]):
            bottom[0] = total
            bottomStation[0] = rowList[-3]
        elif (total <= bottom[1]):
            bottom[1] = total
            bottomStation[1] = rowList[-3]
        elif (total <= bottom[2]):
            bottom[2] = total
            bottomStation[2] = rowList[-3]
    print('{0:s}에서 이용객이 가장 많은 역 순위: '.format(keys[i]))
    print('1. {0:s} ({1:d})\n2. {2:s} ({3:d})\n3. {4:s} ({5:d})'.format(topStation[0], top[0], topStation[1], top[1], topStation[2], top[2]))
    print('{0:s}에서 이용객이 가장 적은 역 순위: '.format(keys[i]))
    print('1. {0:s} ({1:d})\n2. {2:s} ({3:d})\n3. {4:s} ({5:d})\n'.format(bottomStation[0], bottom[0], bottomStation[1], bottom[1], bottomStation[2], bottom[2]))
    top[0] = top[1] = top[2] = 0
    bottom[0] = bottom[1] = bottom[2] = 999999
    topStation[0] = topStation[1] = topStation[2] = bottomStation[0] = bottomStation[1] = bottomStation[2] = ''


file.close()
```
